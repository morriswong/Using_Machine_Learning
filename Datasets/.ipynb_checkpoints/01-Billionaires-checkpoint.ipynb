# Billionaires.

Found out an interesting [post](https://www.bloomberg.com/billionaires/) on Bloomberg listing 500 richest people on earth (publicly), so crawled down the data and did some plots.

As of March 30, 2017

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from Function_01_Billionaires import *
```

```python
bil = pd.read_csv("Data/billionaires.csv", sep = ",")
```

```python
print(bil.head(10))
```

```python
# billionaires' total worth by country & industry

total_Billionaires_net_worth_by_country = {"countries": bil["Country"].unique(),"totalNetWorths": []}
total_Billionaires_net_worth_by_industry = {"industries": bil["Industry"].unique(), "totalNetWorths": []}


for country in total_Billionaires_net_worth_by_country["countries"]:
    total_Billionaires_net_worth_by_country["totalNetWorths"].append(bil["Total_net_worth"].loc[bil["Country"] == country].sum())
for industry in total_Billionaires_net_worth_by_industry["industries"]:
    total_Billionaires_net_worth_by_industry["totalNetWorths"].append(bil["Total_net_worth"].loc[bil["Industry"] == industry].sum())

total_Billionaires_net_worth_by_country = pd.DataFrame(total_Billionaires_net_worth_by_country)
total_Billionaires_net_worth_by_industry = pd.DataFrame(total_Billionaires_net_worth_by_industry)
```

## Total Billionaires Net Worth By Country

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (16, 6)

plot_networth(total_Billionaires_net_worth_by_country)
```

## Total Billionaires Net Worth By Industry

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (16, 6)

plot_networth(total_Billionaires_net_worth_by_industry)
```

### The total money of the Richest 500

## Last Change Combined of Billionaires by Industry

```python
Last_change_in_industries = bil[["Last_change", "Industry"]].groupby("Industry") \
                            .sum().reset_index().sort_values(["Industry", "Last_change"]) \
                            .sort_values("Last_change", ascending=False)

plot_change(Last_change_in_industries)
```

## YTD Change Combined of Billionaires by Industry

```python
YTD_change_in_industries = bil[["YTD_change", "Industry"]].groupby("Industry").sum() \
                           .reset_index().sort_values(["Industry", "YTD_change"]) \
                           .sort_values("YTD_change", ascending=False)

plot_change(YTD_change_in_industries)
```

## Last Change Combined of Billionaires by Country

```python
Last_change_in_country = bil[["Last_change", "Country"]].groupby("Country").sum() \
                         .reset_index().sort_values(["Country", "Last_change"]) \
                         .sort_values("Last_change", ascending=False)

plot_change(Last_change_in_country)
```

```python
YTD_change_in_country = bil[["YTD_change", "Country"]].groupby("Country").sum() \
                        .reset_index().sort_values(["Country", "YTD_change"]) \
                        .sort_values("YTD_change", ascending=False)
        
plot_change(YTD_change_in_country)
```
