# Billionaires.

Found out an interesting [post](https://www.bloomberg.com/billionaires/) on Bloomberg listing 500 richest people on earth (publicly), so crawled down the data and did some plots.

As of March 30, 2017

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from Function_01_Billionaires import *
```

```python
bil = pd.read_csv("Data/billionaires.csv", sep = ",")
```

```python
bil.head(5)
```

## Total Billionaires Net Worth By Country

```python
# count billionaires by country
total_Billionaires_net_worth_by_country = bil.groupby(["Country"])["Total_net_worth"].sum()
total_Billionaires_net_worth_by_country_sorted = total_Billionaires_net_worth_by_country.nlargest(len(total_Billionaires_net_worth_by_country))

plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
total_Billionaires_net_worth_by_country_sorted.plot(kind="bar", label='$1 Trillion')
plt.legend()
plt.title('Total Billionaires Net Worth by Country')
```

## Total Billionaires Net Worth By Industry

```python
# count billionaires by industry
total_Billionaires_net_worth_by_industry = bil.groupby(["Industry"])["Total_net_worth"].sum()
total_Billionaires_net_worth_by_industry_sorted = total_Billionaires_net_worth_by_industry.nlargest(len(total_Billionaires_net_worth_by_industry))

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
total_Billionaires_net_worth_by_industry_sorted.plot(kind="bar", label='$100 Billion')
plt.legend()
plt.title('Total Billionaires Net Worth by Industry')
```

### The total money of the Richest 500

# Number of billionaires in each Country

```python
country_bil_count = bil["Country"].value_counts()

plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
country_bil_count.plot(kind="bar")
plt.title('Number of Billionaires In Each Country')
```

# Number of billionaires in each Industry

```python
industry_bil_count = bil["Industry"].value_counts()

plt.figure(num=None, figsize=(6, 6), dpi=80, facecolor='w', edgecolor='k')
industry_bil_count.plot(kind="bar")
plt.title('Number of Billionaires In Each Industry')
```

# Last Change Combined of Billionaires by Industry

```python
Last_change_in_industries = bil.groupby("Industry")["Last_change"].sum()
Last_change_in_industries_sorted = Last_change_in_industries.nlargest(len(Last_change_in_industries))

plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
Last_change_in_industries_sorted.plot(kind="bar", label='$1 Billion')
plt.legend()
plt.title('Last Change Combined of Billionaires by Industry')
```

# YTD Change Combined of Billionaires by Industry

```python
YTD_change_in_industries = bil.groupby("Industry")["YTD_change"].sum()
YTD_change_in_industries_sorted = YTD_change_in_industries.nlargest(len(YTD_change_in_industries))

plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
YTD_change_in_industries_sorted.plot(kind="bar", label='$10 Billion')
plt.legend()
plt.title('YTD Change Combined of Billionaires by Industry')
```

# Last Change Combined of Billionaires by Country

```python
Last_change_in_country = bil.groupby("Country")["Last_change"].sum()
Last_change_in_country_sorted = Last_change_in_country.nlargest(len(Last_change_in_country))

plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
Last_change_in_country_sorted.plot(kind="bar", label='$1 Billion')
plt.legend()
plt.title('Last Change Combined of Billionaires by Country')
```

# YTD Change Combined of Billionaires by Country

```python
YTD_change_in_country = bil.groupby("Country")["YTD_change"].sum()
YTD_change_in_country_sorted = YTD_change_in_country.nlargest(len(YTD_change_in_country))

plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
YTD_change_in_country_sorted.plot(kind="bar", label='$10 Billion')
plt.legend()
plt.title('YTD Change Combined of Billionaires by Country')
```
